```python
# â€¢	For the same columns Calculate IQR and identify outliers in these columns
for column in float_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identifying outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    print(f"Column: {column}")
    print(f"IQR: {IQR}")
    if not outliers.empty:
        print(f"Number of outliers in {column}: {len(outliers)}")
    else:
        print(f"No outliers found in {column}")
    print("----------")
```

```python
# â€¢	Make Histograms for all Numerical columns and identify the distribution of type by calculating skewness; 


for column in float_columns:
    # Plotting the histogram
    plt.figure(figsize=(10, 6))
    df[column].hist(bins=30)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    
    # Display the plot
    plt.show()
    
    # Calculating skewness
    skewness = df[column].skew()
    print(f'Skewness of {column}: {skewness}')
    if skewness > 0:
        print(f'{column} is positively skewed\n')
    elif skewness < 0:
        print(f'{column} is negatively skewed\n')
    else:
        print(f'{column} is approximately symmetric\n')
```
